```python
students = [
    {"name": "Ali", "note": 14, "filiere": "CS"},
    {"name": "Sara", "note": 11, "filiere": "Math"},
    {"name": "Omar", "note": 9, "filiere": "CS"}
]
```

```python
good_students = [s["name"] for s in students if s["note"] >= 12]
print(good_students)
```

```python
for s in students:
    for key, value in s.items():
        print(key, ":", value)
    print("-----")
```
